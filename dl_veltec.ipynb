* Como possuo como saída quatro classes, precisamos fazer uma transformação desses dados
    * Primeiro passo: Transformar a classe treinamento (formato categórico) para dados numéricos (necessário para a rede neural, pois esta só trabalha com numeros dentro)

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
Classe = label_encoder.fit_transform(classe) #transforma dados categóricos em numéricos
```

```python
Classe
```

* Segundo passo: 
    Como temos 4 neurônios na classe de saída, devemos criar um vetor de 4 dimensões    
    Ou seja, será criado a variável dummy

```python
from keras.utils import np_utils
classe_dummy = np_utils.to_categorical(Classe) 

classe_dummy
```

* Separação entre a base de treino e teste

```python
from sklearn.model_selection import train_test_split
```

```python
#25% para teste e o restante para treinamento
previsores_treinamento, previsores_teste,classe_treinamento,classe_teste = train_test_split(previsores, classe_dummy, test_size=0.25) 
```

* Estrutura da Rede Neural

```python
from keras.models import Sequential 
from keras.layers import Dense #um neuronio conectado a todos os outros
```

```python
classificador = Sequential()
```

```python
'''
Determina a quantidade de entrada somada com a quantidade de saidas (bom, ótimo, regular, ruim) dividido por 2.
'''
qde_neuronios = (84+5)/2 
qde_neuronios
```

* Configuração da primeira camada oculta

```python
'''
neurônios = quantidade de neurônios , ou seja, 48 neurônios (previsores+classes/2)
função de ativação = relu
input_dim = na primeira camada eu preciso dizer quantos neurônios eu tenho na entrada, ou seja,
quantos atributos previsores eu terei
'''

classificador.add(Dense(units = 45, activation = 'relu', input_dim=84)) #input_dim = shape
```

* Configuração da Segunda camada oculta

```python
classificador.add(Dense(units = 5, activation = 'relu'))
```

* Configuração para a camada de saída

Função de ativação:
    * Sigmóide: Função de classificação binária que envolve a classificação de duas variáveis (é ou não é). Gera uma probabilidade para cada neurônio. Pega o maior valor e retorna ele automaticamente.
    * Softmax : Problemas de classificação que envolvem mais de duas classes. Gera uma probabilidade para cada neurônio. Pega o maior valor e retorna ele automaticamente.

```python
#4 neuronios na camada de saída, que é o número de classes que desejo para classificação
#função de ativação para um problema com mais de duas classes: softmax


classificador.add(Dense(units = 5, activation = 'softmax'))
```

```python
from keras import optimizers

optimizer = optimizers.Adam(clipvalue=0.5)
```

* Compilando a rede neural

```python
classificador.compile(optimizer ='Adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
classificador.fit(previsores_treinamento, classe_treinamento, batch_size = 10, epochs = 150)
```

```python
resultado = classificador.evaluate(previsores_teste, classe_teste)
```
