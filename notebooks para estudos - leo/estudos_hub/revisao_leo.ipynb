* Regressao Linear = aproxima um conjunto de dados para uma reta (funcao linear)

```python
from sklearn.linear_model import LinearRegression as lr
from sklearn.datasets import load_iris

lr
```

```python
data_adv = pd.read_csv('Advertising.csv')

data_copy = data_adv.copy(deep=True)

#copia do BD para nao mexer nelaAa
data_copy.drop('Unnamed: 0', axis =1, inplace = True)

data_copy.head()
```

* Separando o dataset

```python
from sklearn.model_selection import train_test_split;


X_train, X_test, y_train, y_test = train_test_split(data_copy.TV, data_copy.sales, test_size = 0.3)
```

```python
plt.scatter(X_train, y_train)
```

```python
lr_1 = LinearRegression()

lr_1.fit(X_train[:,np.newaxis], y_train)



y_pred = lr_1.predict(X_test[:,np.newaxis])

y_pred


print('score:', lr_1.score(y_test[:,np.newaxis], y_pred))
```
